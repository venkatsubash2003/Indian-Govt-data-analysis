```python
#5. How many projects (in percentage) have completed on time?
```

```python
new = df[["Concession End Date(As per agreement)","Concession End Date(Actual)"]]


new.drop(new[new["Concession End Date(Actual)"]=="Not Available"].index,inplace=True)
new.drop(new[new["Concession End Date(Actual)"]=="Not Applicable"].index,inplace=True)
new.drop(new[new["Concession End Date(As per agreement)"]=="Not Available"].index,inplace=True)
new.drop(new[new["Concession End Date(As per agreement)"]=="Not Applicable"].index,inplace=True)

ontime = new[new["Concession End Date(Actual)"] == new["Concession End Date(As per agreement)"]].shape[0]
not_on_time = new[new["Concession End Date(Actual)"] != new["Concession End Date(As per agreement)"]].shape[0]
```

```python
print(ontime,not_on_time)
#8 projects have been completed on time
```

```python
total = ontime + not_on_time
print(total)
```

```python
ontime_pct = (ontime/total) * 100
not_on_pct = 100 - ontime_pct

## A pie chart with one color(green) belongs to completed and another to not completed on time(red) with their percentage
plt.figure(figsize=(14,12))
plt.title('Status of projects',fontsize=14)
plt.pie([ontime_pct,not_on_pct],autopct = '%1.1f%%',labels = ["Ontime finished","Not on time"],colors=['g','r'])
#df['Type of Grant'].value_counts().plot(kind='pie', autopct='%1.1f%%',figsize=(16,12),colors=['m','c','b'],textprops = {"fontsize":22})
plt.legend(fontsize=12)
```

```python
from datetime import datetime
df1=df[['Project Name','Construction Completion Date(Actual)','Construction Completion Date(As per agreement)']]
df1['Construction Completion Date(Actual)']=pd.to_datetime(df1['Construction Completion Date(Actual)'],errors='coerce')
df1['Construction Completion Date(As per agreement)']=pd.to_datetime(df1['Construction Completion Date(As per agreement)'],errors='coerce')
df1['Date_diff']=df1['Construction Completion Date(Actual)']-df1['Construction Completion Date(As per agreement)']
```
